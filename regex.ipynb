```python
# Regex - Regular Expression

# 1. findall() - find all matches

# Goal : find one or more digits
import re
txt=" Order No are 101,102,103,104,105"
findalldigit=re.findall(r"\d+",txt)
print(findalldigit)

```

```python
# 2. search -  find first match only

# group() is used to extract the matched value from the regex result object.

import re
txt1="Revenue will be 50000 45000 89700 each"
searchtxt=re.search(r"\d+",txt1)
print(searchtxt.group())
```

```python
# 3. match - checks only beginning

# group() is used to extract the matched value from the regex result object.

import re
txt2="Python,C,Java are programming languages)"
matchtxt=re.match(r"\w+",txt2)
# for special symbols
matchsym=re.match(r"\W+",txt2)
print(matchsym)
print(matchtxt.group())
```

```python
# 4. sub() - remove text such as remove symbols,html tags etc

import re
txt3="Price is $500 "
subtxt=re.sub(r"\$","",txt3)
# \$ - for endswith
print(subtxt)
# here we removed $ special symbol
```

```python
# 5. split() - split using pattern

import re
txt4="apple,orange;grapes;pineapple,strawberry"
splittxt=re.split(r"[,; ]",txt4)
print(splittxt)
# split if semicolon,comma and space
```

```python
# 6. finditer - return matched objects one - by - one
import re
txt5="balls were bought between 100 to 200 rupees"
finditertxt=re.finditer(r"\d+",txt5)
for i in finditertxt:
  print(i.group())
```

```python
# let's do industry level examples

# 1. Email extraction based on "\S" - for non-space characters
import re
d1="My mail id : devIT@gmail.com"
findmail=re.findall(r"\S+@\S+",d1)
print(findmail)
```

```python
import re
# 2. Phone number extraction - to extract only 10 digits
phone="My phone number is 9876543210"
phonenum=re.findall(r"\d{10}",phone)
print(phonenum)

```

```python
# 3. Remove special characters
import re
splchar ="Hello world@123%9034"
spl=re.sub(r"\W+","",splchar)
print(spl)
```

```python
# 4. Extract dates
import re
dates="My Dob is 19-05-2026"
dob=re.findall(r"\d{2}-\d{2}-\d{4}+",dates)
print(dob)
```

```python
# 5. Extract Hashtags
import re
words="Learn #Python and #Datascience to strong in AI"
extractHashtag=re.findall(r"#\w+",words)
print(extractHashtag)
```

```python
# Customer Name: Jayasree
# Customer Email: sree123@gmail.com
# Customer phone: 9876543210
# Customer dob: 19-05-2026

import re

cust_data="""Customer Name: Jayasree
Customer Email: sree123@gmail.com
Customer phone: 9876543210
Customer dob: 19-05-2026"""

email=re.findall(r"\S+@+\S+.+\S",cust_data)

phone=re.findall(r"\d{10}",cust_data)

dob=re.findall(r"\d{2}-\d{2}-\d{4}",cust_data)
print(email)
print(phone)
print(dob)
```
