```python
import numpy as np
from scipy.stats import kurtosis as k

# data
profit=[120,34,89,9,23,58,99,230,54,80,2,4,5,92]

# find kurtosis using fisherman method
kurt_f=k(profit,fisher=True)
# find kurtosis using without fisherman method
kurt_w=k(profit,fisher=False)

mean=np.mean(profit)
min=np.min(profit)
max=np.max(profit)

if(kurt_f>0):
  print("Heavy tail more distribution")
  if(max-mean)>(mean-min):
    print("More profit")
  else:
    print("Less profit")
elif(kurt_f<0):
  print("Low tail less distribution,so less profit..")
else:
  print("Normal distribution and normal profit/loss")

```

```python
import matplotlib.pyplot as plt
plt.hist(profit,color="purple",edgecolor="black")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit or Loss distribution")
plt.show()
```

```python
# let find mean and median value of profit and plot them as graph

mean_val=np.mean(profit)
median_val=np.median(profit)
print("Mean value for profit:",mean_val)
print("Median value for profit:",median_val)

# graph
plt.plot(profit,color="blue")
plt.axhline(mean_val,linestyle="--",color="green",label="Mean",linewidth=3)
plt.axhline(median_val,linestyle="--",color="red",label="Median",linewidth=3)
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit or Loss distribution")
plt.legend()
plt.show()

```

```python
# Example for Platykurtic - less tail,less variation in data

import numpy as np
from scipy.stats import kurtosis as k

# data
sales=[12,14,17,18,12,20,16,15]

kurto=k(sales,fisher=True)
print("Kurtosis for Sales data:",kurto)

if(kurto<0):
  print("Platykurtic - less tail,less variation in data")
else:
  print("Leptokurtic - more tail,more variation in data")
```

```python
# Example for Mesokurtic - normal distribution

import numpy as np
from scipy.stats import kurtosis as k

# data - Number of fruits in each trees in a garden

fruits=[12,15,18,19,21,24,27,28,29]

kurtos=k(fruits,fisher=True)
print("Kurtosis for fruits in garden:",kurtos)

# since perfect 0 is rare, so we use value just above the zero
if(kurtos<0.1):
  print("Mesokurtic - normal distribution")
else:
  print("Leptokurtic - more tail,more variation in data")
```
