```python
# 1. 50 students, mean = 165, SD = 10
# Find 95% Confidence Interval

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given values
n = 50
mean = 165
sd = 10

# 95% confidence level -> Z value
z = 1.96

# Standard Error
se = sd / np.sqrt(n)

# Confidence Interval
lower = mean - z * se
upper = mean + z * se

print("95% Confidence Interval:")
print(lower, "to", upper)

# ---------------- GRAPH ----------------

x = np.linspace(150, 180, 500)
y = norm.pdf(x, mean, sd)

plt.plot(x, y)

# Confidence interval lines
plt.axvline(lower, linestyle='--')
plt.axvline(upper, linestyle='--')

plt.title("95% Confidence Interval")
plt.xlabel("Student Heights")
plt.ylabel("Density")

plt.show()
```

```python
# 2. Bottle fill mean=500ml, SD=8, n=40; 99% CI.
# Find 99% CI

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# given value
n=40
sd=8
mean=500

# z value for 99% CI
z=2.576
# se -> sd error
se=sd/np.sqrt(n)

# for lower limit
lower=mean-z*se
# for upper limit
upper=mean+z*se
print(lower," to ",upper)

# for graph
x=np.linspace(450,550,500)
y=norm.pdf(x,mean,sd)

plt.plot(x,y)

# confidence interval lines
plt.axvline(lower,linestyle='--')
plt.axvline(upper,linestyle='--')

plt.title("99% Confidence Interval")
plt.xlabel("Bottle Fill")
plt.ylabel("Density")
plt.show()



```
