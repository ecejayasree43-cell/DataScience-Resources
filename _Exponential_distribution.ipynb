```python
# EXPONENTIAL DISTRIBUTION:
# 1. Customer arrives every 10 min; probability within 3 min.


%pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt

mean_time=10
lam=1/mean_time

# prob within 3 mins

# Formula For “within”:  P(X<t)=1−e−λt
t=3

# formula for exponential distribution
prob=np.exp(-lam*t)
print("Exponential distribution result :",prob)
print("In Percentage :",prob*100)

# for generating  random Exponential distribution data
data=np.random.exponential(mean_time,10000)
plt.hist(data,bins=50,density=True,edgecolor='black')

# for 3 mins
plt.axvline(t,color='red',linestyle='--')

plt.xlabel('Time')
plt.ylabel('Probability')
plt.title('Exponential Distribution')
plt.show()


```
