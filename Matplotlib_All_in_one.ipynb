```python
# 1. Line plot -> plt.plot(x,y)
# used to show the trend of the data
# simply, data changes over time
# we use continuous data here like stock price changes, temp changes, website traffic


import matplotlib.pyplot as plt
sales=[130,160,250,450,560,345,890]
months=["Jan","Feb","Mar","Apr","May","Jun","Jul"]
plt.plot(months,sales)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis\n Line Plot Graph")
plt.show()
```

```python
# From analysis, sales is drastically increased on Jul

# Upward line → growth
# Downward line → decrease
# Zig-zag → fluctuations
```

```python
# 2. Bar chart
# It is for discrete data analysis such as comparing 2 categories or 2 quantities

import matplotlib.pyplot as plt
std_Id=[101,102,103,104,105]
marks=[350,450,520,470,590]
plt.bar(std_Id,marks)
plt.title("Student Marks \n Bar Chart")
plt.xlabel("Student Id")
plt.ylabel("Marks")
plt.show()
```

```python
# From bar chart, Student Id: 105 has tall bar , so it scored high marks
```

```python
# 3. Pie chart - It is used for percentage proportions such as market share,vote percentage
# budget allocation and expenses distribution etc

import matplotlib.pyplot as plt
labels=['House Rent','Food','Entertainment','Savings','Shopping','Travel']
salary=[12000,3000,3000,5000,6000,3500]
plt.pie(salary,labels=labels,autopct="%1.1f%%")
plt.title("Salary-Expenses percentage Proportion\n Pie chart")
plt.show()
```

```python
# 4. Scatter plot - to find relationship between 2 variables - finding connectivity b/w both
# Eg: height vs weight, study hours vs marks, salary vs experience

import matplotlib.pyplot as plt
height=[162,150,149,154,172]
weight=[60,45,53,65,70]
plt.scatter(height,weight)
plt.title("Height vs Weight Relationship \n Scatterplot")
plt.show()
```

```python
# # Here in scatterplot,
# The above graph says that taller people have more weight - positive correlation

# Note:
# points moves upward - positive correlation
# points moves downward - negative correlation
# points stands random - no correlation
```

```python
# 5. Histogram - Understanding data distribution, frequency analysis and
# normal distribution

# used for :
# 1. Age or salary distributions
# 2. Exam score distributions, customer purchase frequency etc


# To understand : spread of data, frequency of data,distribution shape

import matplotlib.pyplot as plt
ages=[20,34,80,56,44,44,89,34,30,20,16,20,56,20]
plt.hist(ages,bins=5)
plt.show()
```

```python
# The above graph says large amount of values on left side - Positive skew [Large values] - right skew
# If small amount of values on right side-  Negative skew - left skew

```

```python
#6.Box plot - to show median,outliers or spread over the data

import matplotlib.pyplot as plt
salary=[20000,34000,45000,17000,18900,37000,42000]
plt.boxplot(salary)
plt.title("Salary Distribution\n Box plot")
plt.show()
```

```python
# From this above box plot, it says salary is properly distributed
# it's median = 34000, no outliers present, so perfect spread
```

```python
# 7. Heatmap

# Suppose a teacher wants to see:

# Which students scored HIGH and LOW marks?

# Instead of reading numbers,
# colors help identify quickly.


import matplotlib.pyplot as plt

import numpy as np

marks=np.array([
    [57,89,90],
    [68,80,100],
    [100,74,69]
    ])

plt.imshow(marks)#creates heatmap

subjects=["Maths","Science","Social"]
sections=['A','B','C']

plt.xticks([0,1,2],subjects)
plt.yticks([0,1,2],sections)

plt.title("Students Marks \n Heatmap")
plt.colorbar()
plt.show()
```

```python
# From this heatmap, it shows Student from A, good in Science and social,weak in maths
# llry, Student B is good in social and avg in science and low in maths
# llry, Student C is good in maths and avg in science and low in social

# Heatmap - make data easily understandable instead of manual numerical analysis of data
```
