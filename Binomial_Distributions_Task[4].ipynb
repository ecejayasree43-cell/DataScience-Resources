```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Q1: 95% defect-free items; probability exactly 9 good out of 10.
n=10
p=0.95
x=9

prob=binom.pmf(x,n,p)
print(prob)
```

```python
tot=np.arange(0,n+1)
pmf_values=binom.pmf(tot,n,p)
plt.bar(tot,pmf_values,color='powderblue',edgecolor='black',label="Binomial PMF")
# another bar plot
plt.bar(x,prob,color='powderblue',edgecolor='black',label=f"x={x}")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title(f"Binomial distribution")
plt.legend()
plt.grid(axis="y", alpha=0.4)
plt.show()
```

```python
# but for practical case: we use cdf
# similar data , let see the output below
from scipy.stats import binom
n=10
p=0.95
x=9

p=binom.cdf(x,n,p)
print(p)
```

```python
 # ==========================Question 2:Student solves 60% problems; probability solves at least 3 of 5.==============
```

```python

n1=5
p1=0.60
x1=3

# atleast 3 of 5 so, x1=3, x2=4, x3=5

x2=4
x3=5

xx=x1+x2+x3



prob1=binom.pmf(x1,n1,p1)
prob2=binom.pmf(x2,n1,p1)
prob3=binom.pmf(x3,n1,p1)

probs=prob1+prob2+prob3

print(probs)
```

```python
tot1=np.arange(0,n+1)
pmf_values=binom.pmf(tot1,n1,p1)
plt.bar(tot1,pmf_values,color='pink',edgecolor='black',label="Binomial PMF")
# another bar plot
plt.bar(xx,probs,color='powderblue',edgecolor='red',label=f"xx={xx}")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title(f"Binomial distribution")
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
```
