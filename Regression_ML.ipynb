```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([45,50,55,65,70])

model=LinearRegression()
model.fit(x,y)

y_pred=model.predict(x)

r2=r2_score(y,y_pred)

print("Slope: ",model.coef_[0])
print("Coefficient : ",model.intercept_)
print("R2 Score :",r2)

# prediction for 6 hours
prediction=model.predict([[6]])
print("Prediction for 6 hours be ",prediction[0])

sb.scatterplot(x=x.flatten(),y=y,label="Actual Data")
sb.lineplot(x=x.flatten(),y=y_pred,label="Regression line")

plt.title("Linear Regression Graph")
plt.xlabel("Hours studied")
plt.ylabel("Marks Gained")
plt.legend()
plt.show()

# Linear Regression is a Machine Learning algorithm used to:

# predict one value based on another value using a straight line relationship.
```

```python
# Multiple Linear Regression is used when:

# one output depends on multiple input factors.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from mpl_toolkits.mplot3d import Axes3D

# Dataset
X1 = [1000, 1500, 2000]
X2 = [2, 3, 3]
Y  = [50, 70, 85]


# let's grp data
X=np.column_stack((X1,X2))
Y=np.array(Y)

# model
Model=LinearRegression()
Model.fit(X,Y)

# Prediction line for plane
x1_range = np.linspace(min(X1), max(X1), 10)
x2_range = np.linspace(min(X2), max(X2), 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
y_grid = Model.predict(np.column_stack((x1_grid.ravel(), x2_grid.ravel()))).reshape(x1_grid.shape)

# 3D plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

# Scatter actual points
ax.scatter(X1,X2,Y,color='blue',label="Actual data ")

# Regression plane
ax.plot_surface(x1_grid,x2_grid,y_grid,color='red',alpha=0.5,label="Regression Plane")

ax.set_xlabel('Size (in sqr feets)')
ax.set_ylabel('No of Bedrooms')
ax.set_zlabel('Price in lakhs')
ax.set_title("Multi Linear Regression - 3D view ")
ax.legend()
plt.show()

# Multiple Linear Regression is a Machine Learning algorithm used to:

# predict one value based on multiple input factors.

```

```python
# Logistic Regression

# Logistic Regression predicts categories using probabilities and an S-shaped curve.

# Example: yes/no , pass/fail, spam/not spam

import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns


# dataset

X = np.array([[2], [3], [4], [5], [6], [7]])   # Hours studied
y = np.array([0, 0, 0, 1, 1, 1])               # Fail/Pass

# Model
model = LogisticRegression()
model.fit(X, y)

# Prediction for graph
x_range = np.linspace(1, 8, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_range)[:, 1]

# Model
model = LogisticRegression()
model.fit(X, y)

# Prediction for graph
x_range = np.linspace(1, 8, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_range)[:, 1]

# Accuracy
y_pred = model.predict(X)
acc = accuracy_score(y, y_pred)

print("Accuracy:", acc)
print("Predicted result for 6 hours:", model.predict([[6]])[0])
print("Predicted result for 3 hours:", model.predict([[3]])[0])


plt.figure(figsize=(7, 5))


sns.scatterplot(x=X.flatten(), y=y, label="Actual Data", s=80)


plt.plot(x_range, y_prob, label="Logistic Curve")

plt.xlabel("Hours Studied")
plt.ylabel("Pass Probability")
plt.title("Logistic Regression Curve")
plt.legend()
plt.grid(True)
plt.show()
```
