```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/rainfall1627650742214.csv")
df
```

```python
df.columns
```

```python
annual=df['ANNUAL']
annual
```

```python
from scipy.stats import skew
result=skew(annual)
if result>0:
  print("Positive Skewness")
elif result<0:
  print("Negative Skewness")
else:
  print("Normal Skewness")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.hist(annual,color='pink',bins=10)
plt.xlabel("Annual")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

```python
# tail of the graph lies on positive side- so positive skew (using skewness)
print("Hence, Skewness - to find direction and outliers of the dataset")
```
