```python
# Poisson Distribution is used when we want to find:

# “How many times an event happens in a fixed time or area?”

# Especially when:

# events happen randomly
# events are independent
# and the average occurrence is known.

```

```python
# 1. Website gets 6 visits/min; probability exactly 10 next minute.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=6
x=10 #x-> avg val
pos=p.pmf(x,lam)
print("Prob for getting visits for next 10 mins : ",pos)
xaxis=np.arange(0,21)
fit=p.pmf(xaxis,lam)
plt.bar(xaxis,fit,color="green")
plt.title("Poission Distribution")
plt.xlabel("No of calls per min")
plt.ylabel("Probability")
plt.show()

# Result: Hence 4.13% prob calls for next 10 mins
```

```python
# 2. Bank gets 3 customers/5 min; probability 5 in next 5 min.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=3
x=5 #x-> avg val
pos=p.pmf(x,lam)
print("Prob for getting visits for next 5 mins : ",pos)
xaxis=np.arange(0,10)
fit=p.pmf(xaxis,lam)
plt.bar(xaxis,fit,color="orange")
plt.title("Poission Distribution")
plt.xlabel("No of customers per min")
plt.ylabel("Probability")
plt.show()

# Result: 10.08% prob for 5 customers in next 5 min
```

```python
# 3. Hospital gets 10 patients/hr; probability >12 next hour.

# note: for exact prob  value only we use pmf
# for finding more or less, atmost, atleast,below,upto value -> use cdf

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=10
x=12 #x-> probability >12 next hour
pos=1-p.cdf(x,lam)
# p(X>x)=1- p(X<=x)
print("Prob for getting visits for next 10 mins : ",pos)
xaxis=np.arange(0,21)
fit=p.pmf(xaxis,lam)
plt.bar(xaxis,fit,color="purple")
plt.title("Poission Distribution")
plt.xlabel("No of patients per hours")
plt.ylabel("Probability")
plt.show()

# result: 20.88% prob for getting 12 customers next 1 hr
```

```python
# 4. Call center gets 12 calls/hr; probability no calls in 5 min.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=12*(5/60)
x=0 #x-> avg val
pos=p.pmf(x,lam)
print("Prob for getting no calls for next 5 mins : ",pos)
xaxis=np.arange(0,21)
fit=p.pmf(xaxis,lam)
plt.bar(xaxis,fit,color="blue")
plt.title("Poission Distribution")
plt.xlabel("No of calls per hour")
plt.ylabel("Probability")
plt.show()

# result: 36.87% prob for no calls in 5 min
```

```python
# 5. Toll booth sees 4 cars/min; probability 2 cars in 30 sec.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=4*(30/60)
x=2 #x-> avg val
pos=p.pmf(x,lam)
print("Prob for getting 2 cars within 30 sec : ",pos)
xaxis=np.arange(0,11)
fit=p.pmf(xaxis,lam)
plt.bar(xaxis,fit,color="magenta")
plt.title("Poission Distribution")
plt.xlabel("No of cars  per mins")
plt.ylabel("Probability")
plt.show()

# result: prob for 2 cars in 30 secs = 27.06%
```
