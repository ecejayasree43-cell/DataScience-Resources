```python
import numpy as np
import pandas as pd

df=pd.read_csv("/content/insurance.csv")
df
```

```python
# let take charges column as data

data=df['charges']
data
```

```python
# let test skewness for data

from scipy.stats import skew
ans=skew(data)
if(ans>0):
  print("Positive Skewness")
elif (ans<0):
  print("Negative Skewness")
else:
  print("Normal Skewness")
```

```python
# let plot them as graph(histogram)

import matplotlib.pyplot as plt
plt.hist(data,bins=10,color="cyan",edgecolor="red")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.title("Charges- Frequency Distribution graph")
plt.grid(True)
plt.figure(figsize=(8,5))
plt.show()
```

```python
# Let do for negative skewness
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
prof=[1,3,2,67,89,75,33,78,84]
profits=np.array(prof)

s=skew(profits)
if(s>0):
  print("Positive skew")
elif(s<0):
  print("Negative skew")
else:
  print("Normal skew")
```

```python
plt.hist(profits,bins=10,color="maroon",edgecolor="yellow")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit-Frequency Distribution graph")
plt.grid(True)
plt.figure(figsize=(8,5))
plt.show()
```
