```python
# UNIFORM DISTRIBUTION:
# 1. Lucky draw 1–100; probability 30–60

import numpy as np
import matplotlib.pyplot as plt

start=1
end=100

lower = 30
upper = 60

# random uniform data
data=np.random.uniform(start,end,1000)

# Probability calculation
prob = (upper - lower) / (end - start)

# histogram
plt.hist(data,bins=30,density=True,edgecolor='black')

print("Probability between 30 and 60 :", prob)
print("Probability in percentage :", prob * 100, "%")



plt.axvline(30,color='red',linestyle='--')
plt.axvline(60,color='red',linestyle='--')

plt.title("Uniform Distribution")
plt.xlabel("Lucky Draw value")
plt.ylabel("Probability")
plt.grid(True)
# show plot
plt.show()
```
