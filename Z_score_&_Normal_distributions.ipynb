```python
# Z-score	Formula/calculation used to standardize a value
# Z-value in Normal Distribution
#  The resulting standardized value used in probability tables


from scipy.stats import zscore
data=[70,80,90,100,110,120]
print("Data: ",data)
print("Z-score: ",zscore(data))


# Normalization using z-score means:

# “Transforming data so that mean becomes 0 and standard deviation becomes 1.”
```

```python
# ==============NORMAL DISTRIBUTION PROBLEMS ===========================
# 1. Heights mean=170, SD=10; probability >180.

from scipy.stats import norm
# given data
x=180
mu=170 #mean
sig=10 #sd

# order is important: norm.cdf(x, loc=mean, scale=std)

# for >180 prob => p(X>z)=1-p(X<z)
prob_greater=1-norm.cdf(x,mu,sig)
print("Probability for having >180 height is ",round(prob_greater*100,2),"%")
```

```python
# let's solve the same problem by plotting graph
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu=170 #mean
sig=10 #sd

# To create random set of samples
X=np.random.normal(mu,sig,10000)

#histogram (Bell shaped curve)
plt.hist(X,bins=30,density=True,alpha=0.6)

# plot normal distribution curve
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,1000)
y=norm.pdf(x,mu,sig)
plt.plot(x,y,linewidth=2)

# graph
plt.title("Normal Distribution \n Bell-shaped Curve")
plt.xlabel("Height (in cm)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# 2. IQ mean=100, SD=15; probability 90–110.

from scipy.stats import norm
# given data
x1=90
x2=110

mu=100 #mean
sig=15 #sd

# order is important: norm.cdf(x, loc=mean, scale=std)

# for 90-100 , prob => p(z)=p(X<z)
prob_lower=norm.cdf(x1,mu,sig)
print("Probability for having 90-110 IQ is ",round(prob_lower*100,2),"%")

prob_greater=norm.cdf(x2,mu,sig)
print("Probability for having >180 height is ",round(prob_greater*100,2),"%")

print("Result will be :",round(prob_greater-prob_lower,2))
```

```python
# let's solve the same problem by plotting graph
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x1=90
x2=110

mu=100 #mean
sig=15 #sd

prob_lower=norm.cdf(x1,mu,sig)
prob_greater=norm.cdf(x2,mu,sig)
print("Result will be :",round((prob_greater-prob_lower)*100,2),"%")

# To create random set of samples
X=np.random.normal(mu,sig,10000)

#histogram (Bell shaped curve)
plt.hist(X,bins=30,density=True,alpha=0.6)

# plot normal distribution curve
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,1000)
y=norm.pdf(x,mu,sig)
plt.plot(x,y,linewidth=2)

# shaded region
plt.fill_between(x,y,where=(x>=x1)&(x<=x2),alpha=0.2,color='red')

# graph
plt.title("Normal Distribution \n Bell-shaped Curve")
plt.xlabel("Height (in cm)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
